```python
depth = 0
horizontal_pos = 0

def take_action(direction, value):
    
    global horizontal_pos
    global depth
    global aim
    
    match direction:
        case 'forward':
            print(f"--> Moving forward {value}")
            horizontal_pos += value
            depth +=  (aim *value)
        case 'down':
            print(f"v Increasing aim  {value}")
            aim += value
        case 'up':
            print(f"^ Decreasing aim {value}")
            aim -= value
        case _:
            print("Error. Command not recognized")
    print(f'Depth is now at {depth} , Horizontal position is at: {horizontal_pos} , aim is at: {aim}')
```

```python
# forward X increases the horizontal position by X units
# down X increases the depth by X units
# up X decreases the depth by X units
def parse_commands(command):
    action, value = command.split()
    return action,int(value)
```
