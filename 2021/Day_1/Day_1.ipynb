## Day 1 challenge. Found [here](https://adventofcode.com/2021/day/1) 

### Part 1

```python
import math

increments = 0

# The first number doesn't count as an increment, and so start with infinity to ensure it doesn't get counted
previous = math.inf

with open('data.txt', 'r') as reader:
    for i in reader:
        x = float(i)
        if x > previous:
            increments +=1
            print(f'{i} is greater than {previous}. Increment uppded by 1')
        previous = x

print(f'Final count is {increments}')
    
```

## Part 2

```python
increments = 0

first_left = 0
sec_left = 1

with open('data.txt', 'r') as reader:
    inputs = reader.read().splitlines()
    int_input = list(map(int, inputs))
    
    while sec_left < len(int_input) - 2:
        sum_1 = int_input[first_left] + int_input[first_left+1] + int_input[first_left+2]
        sum_2 = int_input[sec_left] + int_input[sec_left+1] + int_input[sec_left+2]

        if(sum_2 > sum_1):
            increments += 1


        first_left += 1
        sec_left += 1
        
    

print(f'Final count is {increments}')

```
