## Day 1 

```python
# Read in the days input
elves = []
cal_count = 0
most_calories = 0

with open('data/day_1.txt','r') as r:
    for line in r:
        if line == '\n':
            elves.append(cal_count)
            if cal_count > most_calories:
                most_calories = cal_count
            cal_count = 0
        else:
            cal_count += int(line)
            
elves.sort(reverse=True)
    
print(f"most_calories is : {elves[0]}")
```

### Part 2

```python
top_three = elves[0] + elves[1] + elves[2]
print(f"Top three most_calories are : {top_three}")
```

# Day 2

## Part 1

```python
# Them
# A = rock
# B = paper
# C = scissors

# You
# X = rock
# Y = paper
# Z = scissors

data_set = 'data/day_2.txt'

total_points = 0

with open(data_set,'r') as r:
    for line in r:
        result = 0
        your_move_points = 0
        
        
        foo = line.split()
#         match = line.match(r".* .*")
        their_move = foo[0]
        your_move = foo[1]
        match(their_move,your_move):
            #They play rock
            case('A','X'):
                result = 3
            case('A','Y'):
                result = 6
            case('A','Z'):
                result = 0
            
            #They play paper
            case('B','X'):
                result = 0
            case('B','Y'):
                result = 3
            case('B','Z'):
                result = 6
                
            #They play scissors
            case('C','X'):
                result = 6
            case('C','Y'):
                result = 0
            case('C','Z'):
                result = 3
                
#         Look at points for your move
        match(their_move,your_move):      
            #Your choice
            case(_,'X'):
                result += 1
            case(_,'Y'):
                result += 2
            case(_,'Z'):
                result += 3
                
        total_points += result
                
        print(f"total_points: {total_points}")
    
```

## Part 2

X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win.

```python
# Them
# A = rock
# B = paper
# C = scissors

# You
# X = lose
# Y = draw
# Z = win

# scores for
# 1 = rock
# 2 = paper
# 3 = scissors


data_set = 'data/day_2.txt'

total_points = 0

with open(data_set,'r') as r:
    for line in r:
        result = 0
        your_move_points = 0
        
        
        foo = line.split()
#         match = line.match(r".* .*")
        their_move = foo[0]
        your_move = foo[1]
        match(their_move,your_move):
            #They play rock
            case('A','X'):
                result = 3
            case('A','Y'):
                result = 1
            case('A','Z'):
                result = 2
            
            #They play paper
            case('B','X'):
                result = 1
            case('B','Y'):
                result = 2
            case('B','Z'):
                result = 3
                
            #They play scissors
            case('C','X'):
                result = 2
            case('C','Y'):
                result = 3
            case('C','Z'):
                result = 1
                
#         Look at points for your result
        match(their_move,your_move):      
            #Your choice
            case(_,'X'):
                result += 0
            case(_,'Y'):
                result += 3
            case(_,'Z'):
                result += 6
                
        total_points += result
                
    print(f"total_points: {total_points}")
    
```
